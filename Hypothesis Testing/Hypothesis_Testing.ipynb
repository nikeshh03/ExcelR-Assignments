# **Chi-Square Test for Independence**

```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency, chi2
```

```python
print("Hypothesis Statement:")
print("Null Hypothesis (H0): There is no association between device type and customer satisfaction (they are independent).")
print("Alternative Hypothesis (Ha): There is a significant association between device type and customer satisfaction.\n")
```

```python
data = np.array([
    [50, 70],   # Very Satisfied
    [80, 100],  # Satisfied
    [60, 90],   # Neutral
    [30, 50],   # Unsatisfied
    [20, 50]    # Very Unsatisfied
])
chi2_stat, p_val, dof, expected = chi2_contingency(data)
```

```python
print("--- Calculations ---")
print(f"Chi-Square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_val:.4f}")
print(f"Degrees of Freedom: {dof}")
```

```python
alpha = 0.05
critical_value = chi2.ppf(1 - alpha, dof)
print(f"Critical Value (alpha=0.05, df={dof}): {critical_value:.4f}\n")
```

```python
print("--- Decision ---")
if chi2_stat > critical_value:
    print("Decision: Reject the Null Hypothesis.")
else:
    print("Decision: Fail to reject the Null Hypothesis.")


print("\n--- Conclusion Report ---")
if chi2_stat > critical_value:
    print("There is a significant association between the type of smart home device purchased and the customer's satisfaction level.")
else:
    print("There is no significant evidence to suggest an association between the type of smart home device and customer satisfaction level.")
```

# **Hypothesis Testing**

```python
import numpy as np
from scipy import stats

print("Hypothesis Statement:")
print("Null Hypothesis (H0): μ <= 4000 (The mean weekly operating cost is not higher than the theoretical model).")
print("Alternative Hypothesis (Ha): μ > 4000 (The mean weekly operating cost is higher than the theoretical model).\n")
```

```python
n = 25
sample_mean = 3050
theoretical_mu = 1000 + (5 * 600)
sigma = 5 * 25
alpha = 0.05

z_stat = (sample_mean - theoretical_mu) / (sigma / np.sqrt(n))
```

```python
print("--- Calculations ---")
print(f"Theoretical Mean (μ): {theoretical_mu}")
print(f"Sample Mean (x̄): {sample_mean}")
print(f"Standard Deviation (σ): {sigma}")
print(f"Standard Error: {sigma / np.sqrt(n)}")
print(f"Test Statistic (Z): {z_stat:.4f}")

critical_z = stats.norm.ppf(1 - alpha)
print(f"Critical Value (Z-table, alpha=0.05): {critical_z:.4f}\n")
```

```python
print("--- Decision ---")
if z_stat > critical_z:
    print("Decision: Reject the Null Hypothesis.")
else:
    print("Decision: Fail to reject the Null Hypothesis.")


print("\n--- Conclusion Report ---")
if z_stat > critical_z:
    print("There is strong evidence to support the claim that weekly operating costs are higher than the model suggests.")
else:
    print("There is no evidence to support the claim that the weekly operating costs are higher. In fact, the sample mean observed is lower than the theoretical model.")
```
