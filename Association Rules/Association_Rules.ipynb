```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
# Load the dataset (no header since the first row contains data)
df = pd.read_csv('/content/Online retail.csv', header=None)
```

```python
# Convert the dataframe into a list of transactions
transactions = []
for i in range(len(df)):
    # Split the comma-separated string and remove whitespace
    row = str(df.values[i, 0]).split(',')
    transactions.append([item.strip() for item in row])
```

```python
# One-hot encoding for association rule mining
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
# Model 1: Support = 0.01, Confidence = 0.2
frequent_itemsets1 = apriori(df_encoded, min_support=0.01, use_colnames=True)
rules1 = association_rules(frequent_itemsets1, metric="confidence", min_threshold=0.2)

# Model 2: Support = 0.005, Confidence = 0.3
frequent_itemsets2 = apriori(df_encoded, min_support=0.005, use_colnames=True)
rules2 = association_rules(frequent_itemsets2, metric="confidence", min_threshold=0.3)

# Model 3: Support = 0.01, Lift = 2
rules3 = association_rules(frequent_itemsets1, metric="lift", min_threshold=2.0)
```

```python
print(f"Number of rules in Model 1: {len(rules1)}")
print(f"Number of rules in Model 2: {len(rules2)}")
print(f"Number of rules in Model 3: {len(rules3)}")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(rules1['support'], rules1['confidence'], alpha=0.5, c='blue')
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.title('Support vs Confidence (Model 1)')
plt.grid()
plt.show()
```

```python
# Displaying top 10 rules based on lift
top_rules = rules1.sort_values('lift', ascending=False).head(10)
print("\nTop 10 Rules by Lift:")
print(top_rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']])
```
